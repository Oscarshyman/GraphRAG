```python
import pandas as pd

node_file="/Users/chenruidong/Documents/Graph Rag/ragtest/output/create_final_nodes.parquet"
entities_file="/Users/chenruidong/Documents/Graph Rag/ragtest/output/create_final_entities.parquet"
relationships_file="/Users/chenruidong/Documents/Graph Rag/ragtest/output/create_final_relationships.parquet"
communities_file="/Users/chenruidong/Documents/Graph Rag/ragtest/output/create_final_communities.parquet"
reports_file="/Users/chenruidong/Documents/Graph Rag/ragtest/output/create_final_community_reports.parquet"
text_units_file="/Users/chenruidong/Documents/Graph Rag/ragtest/output/create_final_text_units.parquet"
documents_file="/Users/chenruidong/Documents/Graph Rag/ragtest/output/create_final_documents.parquet"

nodes_df = pd.read_parquet(node_file)
entities_df = pd.read_parquet(entities_file)
relationships_df = pd.read_parquet(relationships_file)
communities_df = pd.read_parquet(communities_file)
reports_df = pd.read_parquet(reports_file)
text_units_df = pd.read_parquet(text_units_file)
documents_df = pd.read_parquet(documents_file)
```

```python
entities_selected = entities_df[['id', 'title', 'type', 'description']]
entities_selected = entities_selected.rename(columns={"id": "entity_ids"})

entities_selected
```

```python
communities_selected = communities_df[["community","parent","level","entity_ids","relationship_ids","size"]]
communities_selected
```

# relationship between communities and entities

```python
import pandas as pd
import numpy as np

entities_selected = entities_selected.rename(columns={"id": "entity_ids"})

# Convert 'entity_ids' column from list-like format to individual rows
if isinstance(communities_selected['entity_ids'].iloc[0], (list, tuple, np.ndarray)):
    communities_selected = communities_selected.explode('entity_ids')

entities_selected["entity_ids"] = entities_selected["entity_ids"].astype(str)
communities_selected["entity_ids"] = communities_selected["entity_ids"].astype(str)


entities_communities = entities_selected.merge(
    communities_selected, left_on="entity_ids", right_on="entity_ids", how="left"
)

print(entities_communities)
#entities_communities.to_csv("~/Desktop/entities_communities_table.csv", index=False)
```

```python
relationships_selected = relationships_df[['id','source', 'target', 'description', 'weight']]
relationships_selected = relationships_selected.rename(columns={"id":"relationship_ids"})
relationships_selected
```

# relationship between communities and relationship

```python
if isinstance(communities_selected['relationship_ids'].iloc[0], (list, tuple, np.ndarray)):
    communities_selected = communities_selected.explode('relationship_ids')

relationships_selected["relationship_ids"] = relationships_selected["relationship_ids"].astype(str)
communities_selected["relationship_ids"] = communities_selected["relationship_ids"].astype(str)


relationships_communities = relationships_selected.merge(
    communities_selected, left_on="relationship_ids", right_on="relationship_ids", how="left"
)

print(relationships_communities)
relationships_communities.to_csv("~/Desktop/relationships_communities_table.csv", index=False)
```
