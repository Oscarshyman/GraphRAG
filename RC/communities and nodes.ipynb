# find the difference between graphrag4 and graphrag8
# graphrag4 acts as baseline, I increase minimal example numbers from 3 to 4, all other parameters do not change

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
```

# entities title

```python
df4=pd.read_parquet("/Users/chenruidong/Documents/Graph Rag/ragtest4/output/create_final_entities.parquet")
df8=pd.read_parquet("/Users/chenruidong/Documents/Graph Rag/ragtest8/output/create_final_entities.parquet")
print(len(df4["id"]))
print(len(df8["id"]))
```

```python
title4=set(df4["title"].unique())
title8=set(df8["title"].unique())
```

```python
only_in_4=title4-title8
intersection=title4&title8
only_in_8=title8-title4
print(len(only_in_4))
print(len(intersection))
print(len(only_in_8))
```

```python
#!pip install matplotlib-venn
venn2([title4,title8],set_labels=("ragtest4","ragtest8"))
plt.title("Unique vs Shared Titles in Entities")
plt.show()
```

```python
only_in_4
```

# relationships change between ragtest4 and ragtest8

```python
df4_relationship=pd.read_parquet("/Users/chenruidong/Documents/Graph Rag/ragtest4/output/create_final_relationships.parquet")
df8_relationship=pd.read_parquet("/Users/chenruidong/Documents/Graph Rag/ragtest8/output/create_final_relationships.parquet")
# Convert to sets of undirected edges (sorted tuples)
relationship4=set(tuple(sorted([i,j])) for i,j in zip(df4_relationship["source"],df4_relationship["target"]))
relationship8=set(tuple(sorted([i,j])) for i,j in zip(df8_relationship["source"],df8_relationship["target"]))
only_in_4_relationship=relationship4-relationship8
only_in_8_relationship=relationship8-relationship4
in_both_relationship=relationship4 & relationship8
```

```python
only_in_4_relationship
```

```python
venn2([relationship4,relationship8],set_labels=("ragtest4","ragtest8"))
plt.title("Unique vs Shared Titles in relationships")
plt.show()
```
